#### K-Means clustering

Works by assigning data points to clusters based on the shortest distance to the centroids or centre of the cluster, its main goal is to minimise the sum of distances between data points and their respective clusters (REGUNATH, 2022).

```python
x = clu.iloc[:,2:4]
```

```python
x
```

```python
x = clu.iloc[:,2:4].values
```

Since we already know that we have 7 obesity lebels (clusters), we can put our Height and Weight features directly into the K-Means algorithm, these are the results:

```python
kmeans = KMeans(n_clusters=7, max_iter=300, n_init=10, random_state=4)

y_kmeans = kmeans.fit_predict(x)
```

```python
print(sns.color_palette('Set2').as_hex())
```

```python
sns.scatterplot(x=x[y_kmeans==0,1], y=x[y_kmeans==0,0], s=30, c='#e78ac3', label='Obesity_Type_I')
sns.scatterplot(x=x[y_kmeans==1,1], y=x[y_kmeans==1,0], s=30, c='#66c2a5', label='Normal_Weight')
sns.scatterplot(x=x[y_kmeans==2,1], y=x[y_kmeans==2,0], s=30, c='#e5c494', label='Obesity_Type_III')
sns.scatterplot(x=x[y_kmeans==3,1], y=x[y_kmeans==3,0], s=30, c='#fc8d62', label='Overweight_Level_I')
sns.scatterplot(x=x[y_kmeans==4,1], y=x[y_kmeans==4,0], s=30, c='#a6d854', label='Insufficient_Weight')
sns.scatterplot(x=x[y_kmeans==5,1], y=x[y_kmeans==5,0], s=30, c='#ffd92f', label='Obesity_Type_II')
sns.scatterplot(x=x[y_kmeans==6,1], y=x[y_kmeans==6,0], s=30, c='#8da0cb', label='Overweight_Level_II')

plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,0], s=45, c='black', label='Centroids')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend(loc='upper right', bbox_to_anchor=(1.38,1.02));
```

```python
print('K-Means clustering: ', silhouette_score(x,y_kmeans))
```

#### DBSCAN

It determines clusters based on how dense regions are, it can find irregular shaped clusters and outliers quite well (REGUNATH, 2022).

We will search for the maximum distance between two samples for one to be considered as in the neighbourhood, then, we will create a function to iterate different parameters to find the best ones, measuring them with the Silhouette score.

```python
neigh = NearestNeighbors(n_neighbors=10)
nbrs = neigh.fit(clu[['Height','Weight']])

distances, indices = nbrs.kneighbors(clu[['Height','Weight']])
```

```python
distances = distances[:,1]
plt.figure(figsize = (14,5))
plt.plot(distances)
plt.title('K-distance Graph', fontsize = 20)
plt.xlabel('Data Points sorted by distance', fontsize = 14)
plt.ylabel('Epsilon', fontsize = 14)
plt.show()
```

As shown above, we can notice that the epsilon for our data would be between 0.1 and 0.5.

```python
db_parameters = [[0.1,5],[0.2,5],[0.3,5],[0.4,5],[0.5,5],[0.1,4],[0.2,4],[0.3,4],[0.4,4],[0.5,4],
                [0.1,6],[0.2,6],[0.3,6],[0.4,6],[0.5,6],[0.1,7],[0.2,7],[0.3,7],[0.4,7],[0.5,7]]
for ep,min_sample in db_parameters:
    db = DBSCAN(eps=ep, min_samples=min_sample)
    db_clusters = db.fit_predict(x)
    print('Eps: ', ep, 'Min Samples: ', min_sample)
    print('DBSCAN clustering: ', silhouette_score(x,db_clusters))
```

```python
db_clusters.max()
```

From the above function, we can see that the highest Silhouette score (0.4840) is with the eps:0.2 with min samples:4, now with can pass these parameters through the DBSCAN algorithm.

Comparing the K-Means and DBSCAN Silhouette scores, we can expect that DBSCAN will not be doing as well as the K-Means algorithm.


```python
dbscan = DBSCAN(eps=0.2, min_samples=4)
dbscan.fit(clu[['Height','Weight']])
```

```python
plt.figure(figsize = (8, 5))
plt.scatter(x=clu['Weight'], y=clu['Height'], c = db_clusters, cmap = 'tab20b', s = 30)
plt.title('DBSCAN Clustering', fontsize = 20)
plt.xlabel('Weight', fontsize = 14)
plt.ylabel('Height', fontsize = 14)
plt.show()
```

As we can observe, based on the Silhouette score, we can say that our assumption that K-Means works better for our data is correct.

Now we need to visualise the Autocorrelation and Partial Autocorrelation of our data.

Autocorrelation (ACF): Calculation of the correlation of the time series observations with values of the same series, but at previous times.

Partial Autocorrelation (PACF): Summarizes the relationship between an observation in a time series with observations at previous time steps, but the relationships of intervening observations are removed.

(Marv, 2021)


```python
plot_time_s(time_s['Close'], nlag = 100)
```

From these plots, we are looking at the points outside the blue shadow, that means that they have statistical significance, that indicates that an observed relationship is unlikely to be due to chance.

Since our Autocorrelation plot have more than 20 lags with positive correlation above 0.5, we can expect that we will need a differencing above 1.


Let’s see if our data is stationary or not by performing a Dickey Fuller test.

```python
adf_result = adfuller(time_s.Close.to_list())
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' % adf_result[1])
```

In this case, we will use the ARIMA algorithm, this algorithm is used to understand past data or predict future data in a series.
To search for the best values for our ARIMA algorithm, we will create a function to iterate multiple values through the algorithm, we are looking for the lowest Akaike’s Information Criterion (AIC)


```python
# Search for the best AIC values
aicValues = []

for d in range(1,3):
    for ari in range(0,3):
        for maj in range(0,3):
            try:
                arima_obj = ARIMA(time_s['Close'].tolist(), order=(ari,d,maj))
                arima_obj_fit = arima_obj.fit()
                aicValues.append([ari, d, maj, arima_obj_fit.aic])
                
            except ValueError:
                pass

print(aicValues)
                
```

Best AIC value: [0, 2, 2, 3585.3853420221676]

```python
arima_obj = ARIMA(time_s['Close'].tolist(), order=(0,2,2))

arima_obj_fit = arima_obj.fit()

arima_obj_fit.summary()
```

```python
pred = arima_obj_fit.fittedvalues

time_s['ARIMA'] = pred
diffval = arima_obj_fit.resid + arima_obj_fit.fittedvalues
time_s['diffval'] = diffval
time_s
```

```python
fig, ax = plt.subplots(figsize=(8,6))

residuals = arima_obj_fit.resid[2:]

sm.qqplot(residuals, line='s', ax=ax)
ax.set_title('Q-Q plot, quantiles of X vs quantiles of a distribution', fontdict = {'fontsize':18});
plt.show()
```

```python
ax_f, axarr = plt.subplots(1, sharex=True);
ax_f.set_size_inches(14, 6);

time_s['diffval'].iloc[2:].plot(color='b', linestyle='-', ax=axarr, label='Actual');
time_s['ARIMA'].iloc[2:].plot(color='r', linestyle='--', ax=axarr, label='ARIMA Forecast');
axarr.set_title('ARIMA (0,2,2)', fontdict = {'fontsize':18});
axarr.set_xlabel('Index');
axarr.set_ylabel('Closing value');
axarr.legend();
```

As we can observe, the ARIMA algorithm works efficiently with our dataset and we can see that by looking at the confidence intervals.

```python
result = arima_obj_fit.get_forecast()
forecast = result.predicted_mean

# summarize confidence intervals
intervals = [0.2, 0.1, 0.05, 0.01]

for a in intervals:
    ci = result.conf_int(alpha=a)
    print('%.1f%% Confidence Interval: %.3f between %.3f and %.3f' % ((1-a)*100, forecast, ci[0,0], ci[0,1]))
```

In the case of the 95% confidence interval, suggests that there is a high likelihood that the real observation will be within the range of 432.637 and 474.824.
